```python
import requests
from bs4 import BeautifulSoup
page = requests.get("http://104.238.111.129:92/SWTInterview/Home/List?from=&todate=&candidatename=Select&companyname=Select&keyword=python&Area=select")
soup = BeautifulSoup(page.content,'html.parser')
soup
```

```python
my_table = soup.findAll('table', {'class':'table table-hover table-striped'})
my_table
```

```python
res_map = {'Candidate_Name':[], 'Email':[], 'Phone':[], 'Company_Name':[], 'Interview_Date': [],\
          'Area':[], 'Description':[]}
for i in range(len(my_table[0].findAll('td'))//7):
    res_map['Candidate_Name'].append(my_table[0].findAll('td')[i*7].renderContents().strip().decode("utf-8"))
    res_map['Email'].append(my_table[0].findAll('td')[i*7+1].renderContents().strip().decode("utf-8"))
    res_map['Phone'].append(my_table[0].findAll('td')[i*7+2].renderContents().strip().decode("utf-8"))
    res_map['Company_Name'].append(my_table[0].findAll('td')[i*7+3].renderContents().strip().decode("utf-8"))
    res_map['Interview_Date'].append(my_table[0].findAll('td')[i*7+4].renderContents().strip().decode("utf-8"))
    res_map['Area'].append(my_table[0].findAll('td')[i*7+5].renderContents().strip().decode("utf-8"))
    res_map['Description'].append(my_table[0].findAll('td')[i*7+6].renderContents().strip().decode("utf-8"))
```
